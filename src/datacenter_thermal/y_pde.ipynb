```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

# 创建数据集
x = torch.linspace(-1, 1, 1000).view(-1, 1).requires_grad_()
y = 2 * x + torch.randn(1000, 1).requires_grad_() * 0.3  # 生成带有噪声的线性数据

# 定义神经网络模型
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc = nn.Linear(1, 1)  # 一个输入节点和一个输出节点

    def forward(self, x):
        return self.fc(x)

net = Net()

# 定义损失函数和优化器
criterion = nn.MSELoss()
optimizer = optim.SGD(net.parameters(), lr=0.01)

# 创建数据加载器
dataset = TensorDataset(x, y)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

# 训练模型
for epoch in range(100):
    for batch_x, batch_y in dataloader:
        optimizer.zero_grad()
        output = net(batch_x)
        y_grad = torch.autograd.grad(output, batch_x, grad_outputs=torch.ones_like(output), create_graph=True)[0]
        print(y_grad)
        loss = criterion(output, batch_y)
        loss.backward()
        optimizer.step()

# 使用模型进行预测
test_x = torch.tensor([[0.5]])
test_x0 = torch.tensor([[0.5+0.00001]])
test_y = net(test_x)
test_y0 = net(test_x0)
print(test_y.item())  # 输出 f'(x)
print((test_y0.item()-test_y.item())/0.00001)  # 输出 f''(x)

```
